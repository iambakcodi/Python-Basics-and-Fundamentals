```python
#int-float-string(numbered)-bool
var=3#int
float_var=float(var) #float
print(float_var)
str_var=str(var)
print(type(str_var),str_var)
bool_var=bool(var)
print(bool_var)

bool_var2=True
bool_var22=False
print(float(bool_var2))
print(float(bool_var22))
```

# Concept of Hashable Data Type

```python
# 1. Numeric_type #numeric data
# int:                       Whole Numbers without decimal point
# float:                     Decimals

# 2. Set_Type #unordered
# set {a,b,c,d}:             Unordered collection of unique elements. Duplicates are automatically removed.
#                              Elements must be hashable(value not change, so cant put lists/dicts inside it)

# 3. Sequence_Types #ordered
# list [a,b,[c,d],e,'fgh']:  Ordered and mutable sequence.It can contain elements of different data types.
#                              Elements can be of any type(no hashable condition)
                             
# str "krishang":            Ordered and immutable sequence of characters(can contain letters, numbers, symbols, and whitespace)
#                            (') or double quotes (")  
                            
# tuple (a,b,[c,d],e,'fgh'): Ordered and immutable collection of elements. It can contain elements of different data types.

# 4. Mapped_Type #Key Value pairs
# dict :                     Unordered and mutable collection of key(unique)-value pairs.
#                              Keys must be hashable(value not change, so cant put lists/sets as keys)

# 5. Boolean_Type #Truth values and Logic Gates
# bool T/F:                  The truth values True or False. It is often used in conditions and logical operations.

# 6. None_Type #value absence
# None:                      Absence of a value or a null value. 
#                            Signify that a variable or expression has no meaningful value.

# In Python, hashable(value cant change) objects are those that have a hash value that never changes during their lifetime and can 
# be compared to other objects for equality. 
# Mutable objects, which can be changed after creation,are generally not hashable.
# Lists are mutable, so their hash value can change if the contents are modified. So not hashable
# Dictionaries: Dictionaries are also mutable, and their hash value can change if items are added, removed, or modified. So not 
# hashable.
# Sets: Sets are mutable, and their hash value can change if elements are added or removed. So not hashable

# unhashable_set = {1, 2, 3}
# hash(unhashable_set)  # Raises TypeError
```
